На базе Ледового дворца организуются также сборы спортсменов по шорт-треку из других регионов.

> Для начала импортируем необходимые библиотеки для анализы и визуализации имеющихся данных.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

> Для загрузки .csv файла с нужными нам данными поспользуемся функцией read_csv().

```python
df = pd.read_csv("Для корпоративных.csv", sep=';', encoding='windows-1251')
df
```

```python
splot = sns.barplot(x='Стоимость услуги',y='Предоставляемая услуга',data=df, hue= "Спортивный комплекс", palette="Set1")
for i in splot.containers:
    splot.bar_label(i,label_type='center') 
plt.xlabel('')
plt.ylabel('')
plt.xticks (size=10)
plt.yticks (size=10)
```
