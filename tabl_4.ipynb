Создадим в ручную дата фрейм

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'2020': [100, 50, 80, 1000, 1900, 120, 250],
                   '2023': [150, 75, 150, 1800, 3000, 200, 300]}, index=['Билет на массовое катание для взрослых','Льготный билет на массовое катание',
'Прокат коньков', 'Абонемент на 1 месяц без проката коньков', 
'Абонемент на 1 месяц с прокатом коньков', 'Заточка коньков', 'Разогрев 1 пары фигурных и хоккейных ботинок в термопечи'])
df
```

```python
df['Абсолютное изменение'] = pd.Series(df['2023'] - df['2020'])
df
```

```python
df['Темп роста, %'] = np.round((df['2023'] / df['2020']) * 100, 2)
df
```
