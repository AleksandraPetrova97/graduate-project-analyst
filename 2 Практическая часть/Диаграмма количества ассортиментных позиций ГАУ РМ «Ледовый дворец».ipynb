Для начала работы с данными импортируем библиотеки. Осуществим подключение с помощью оператора import.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Создадим DataFrame, передав информацию через словарь, который содержит ключ для каждого столбца и массив значений для них.

```python
data_assortment_items = {'наименование услуги': ['фитнес', 'тренажерный зал', 'ледовая арена', 'солярий', 'прочие услуги'],
                 'количество ассортиментных позиций': [12, 8, 22, 3, 8]}
dframe = pd.DataFrame(data_assortment_items)
dframe
```

Что бы красиво визуализировать данные, воспользуемся библиотеку matplotlib. Для  создания диаграммы, используем данные DataFrame, о количестве ассортиментных позиций предоставляемых услуг ГАУ РМ «Ледовый дворец».
Создадим функцию, которая cможет добавbnm метки значений на столбчатую диаграмму, для этого используем функцию plt.text(). Для добавления меток в центр высоты столбца, разделим координаты y [i] //2. Установим параметр ha, для выравнивания меток значений по центру полосы. 

Зададим размер фигуры figsize, передав этот параметр в функцию создания фигуры plt.figure. Далее создадим одномерный массив данных, в котором будут храниться цвета, какими окрасится диаграмма.

Построим саму диаграмму, используя функцию plt.bar(). Передадим внутрь функции все необходимые параметры. И уже после создания диаграммы, вызовем функцию для добавления меток значений, установим заголовки, метки по оси X и по оси Y диаграммы.

Визуализируем график с помощью функции plt.show() .

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]//2, y[i], ha = 'center')

plt.figure(figsize=(8,4))
colors = ['royalblue', 'red', 'green', 'magenta', 'yellow']
plt.bar(dframe['наименование услуги'], dframe['количество ассортиментных позиций'], color = colors)

addlabels(dframe['наименование услуги'], dframe['количество ассортиментных позиций'])
plt.xticks(rotation=30)
plt.yticks()
plt.xlabel("наименование услуги")
plt.ylabel("Количество ассортиментных позиций, ед")

plt.show()
```
