```python
import pandas as pd
import numpy as np
# импортировали data frame
df = pd.read_csv("Для корпоративных.csv", sep=';', encoding='windows-1251')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = df.groupby('Спортивный комплекс')['Стоимость услуги'].sum().reset_index()
data
plt.figure(figsize = (10,8))
ax = sns.barplot(x = data['Спортивный комплекс'], 
            y = data['Стоимость услуги'],
             palette='hls',
             data=data)

ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)             
```
