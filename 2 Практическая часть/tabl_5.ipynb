```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'2020': [1102700, 731200, 128500, 640200, 646100, 1000, 1263800, 603700],
                   '2021': [1643000, 1081600, 185200, 950200, 961200, 2000, 1842100, 1639500],
                   '2022': [1429400, 930100, 161500, 807600, 855400, 1000, 1639500, 789000]}, index=['массовое катание', 'прокат коньков',
                    'заточка коньков', 'Тренажерный зал', 'Фитнес', 'Солярий',  'хоккей (продажа  билетов)', 'реклама'])
df
```

```python
df['Абсолютное изменение 2021г. к 2020г.'] = pd.Series(df['2021'] - df['2020'])
df['Абсолютное изменение 2022г. к 2021г.'] = pd.Series(df['2022'] - df['2021'])
df['Абсолютное изменение 2022г. к 2020г.'] = pd.Series(df['2022'] - df['2020'])
df
```

```python
df['Темп роста, 2021 г. к 2020 г. %'] = np.round((df['2021'] / df['2020']) * 100, 1)
df['Темп роста, 2022 г. к 2021 г. %'] = np.round((df['2022'] / df['2021']) * 100, 1)
df['Темп роста, 2022 г. к 2020 г. %'] = np.round((df['2022'] / df['2020']) * 100, 1)
df
```

```python
plt.figure(figsize=(10,6))
plt.plot (df.index, df['2020'], label='sales',  marker='D')
plt.plot (df.index, df['2021'], label='sales',  marker='D')
plt.plot (df.index, df['2022'], label='sales', marker='D')

plt.legend(["2020", "2021", "2022"], loc="lower right")
plt.xticks (rotation= 30)
plt.yticks([])
```
