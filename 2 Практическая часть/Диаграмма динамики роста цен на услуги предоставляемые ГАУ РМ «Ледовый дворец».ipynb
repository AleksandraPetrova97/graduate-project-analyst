```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
price_growth_dynamics = pd.read_csv("Приложение_1\Приложение_1_Динамика роста цен на услуги предоставляемые ГАУ РМ «Ледовый дворец».csv", 
                                        sep=';', encoding='windows-1251')
```

```python
price_growth_dynamics.sample(5)
```

```python
price_growth_dynamics.groupby('Год предоставления услуги').count()
```

```python
# задаем цвета диаграммы
colors = ['violet', 'aqua']
# создаем фигуру и задаем её размер
plt.figure(figsize=(10,4))
# задаем данные диаграммы
splot= sns.barplot(x='Стоимость услуги', y='Наименование услуги',
            hue= "Год предоставления услуги", data = price_growth_dynamics, palette='Set2')
# добавляем метки данных
plt.bar_label(splot.containers[0],size=10, label_type='center')
plt.bar_label(splot.containers[1],size=10, label_type='center')
#  подписываем x и y
plt.xlabel('Наименование услуги')
plt.ylabel('Средняя стоимость услуг, руб')

plt.show()
```
