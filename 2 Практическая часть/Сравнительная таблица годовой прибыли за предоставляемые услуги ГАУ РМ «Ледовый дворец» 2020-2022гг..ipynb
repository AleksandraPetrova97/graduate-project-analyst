```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_annual_profit = pd.read_csv("Приложение_1\Приложение_1_Годовая прибыль на предоставляемые услуги ГАУ РМ «Ледовый дворец» 2020-2022гг.csv", 
                                    sep=';', encoding='windows-1251')
df_annual_profit
```

```python
df_annual_profit.info()
```

```python
df = df_annual_profit.pivot_table(
               index=["наименование услуги"],
               columns=["год"],
               values=["годовая прибыль"])
df
```

```python
df['Абс. изм. 2021г. к 2020г.'] = (df['годовая прибыль', 2021] - df['годовая прибыль', 2020])
df['Абс. изм. 2022г. к 2021г.'] = (df['годовая прибыль', 2022] - df['годовая прибыль', 2021])
df['Абс. изм. 2022г. к 2020г.'] = (df['годовая прибыль', 2022] - df['годовая прибыль', 2020])
```

```python
df['Темп роста, 2021 г. к 2020 г. %'] = np.round((df['годовая прибыль', 2021] / df['годовая прибыль', 2020]) * 100, 1)
df['Темп роста, 2022 г. к 2021 г. %'] = np.round((df['годовая прибыль', 2022] / df['годовая прибыль', 2021]) * 100, 1)
df['Темп роста, 2022 г. к 2020 г. %'] = np.round((df['годовая прибыль', 2022] / df['годовая прибыль', 2020]) * 100, 1)
df
```
